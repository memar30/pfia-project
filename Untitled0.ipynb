```python
from bs4 import BeautifulSoup
import requests, json, time
from urllib.parse import urljoin
from dateutil import parser as dateparser
import pandas as pd

COUNTY_CONFIG = {

    "name": "Harris",
    "state": "TX",
    "base_url": "https://example.county.clerk",   # base for resolving relative links
    "start_url": "https://example.county.clerk/search?doctype=lis_pendens&page=1",
    "page_param": "page",
    # Generic selectors — likely need edits per county. I will create exact selectors if you paste your county URL.
    "results_selector": "table.results > tbody > tr",
    "cols": {
        "date": "td:nth-of-type(1)",
        "doc_number": "td:nth-of-type(2)",
        "parties": "td:nth-of-type(3)",
        "address": "td:nth-of-type(4)",
        "doc_link": "td:nth-of-type(5) a"
    }
}

HEADERS = {"User-Agent":"pfia-agent/0.1 (+https://example.com)"}

def fetch_page(url):
    print("Fetching:", url)
    resp = requests.get(url, timeout=30, headers=HEADERS)
    resp.raise_for_status()
    return resp.text

def parse_row(row, cfg):

    def t(sel):
        el = row.select_one(sel)
        return el.get_text(strip=True) if el else None
    def a(sel, attr='href'):
        el = row.select_one(sel)
        return el.get(attr) if el else None

    date_raw = t(cfg['cols']['date'])
    try:
        date = str(dateparser.parse(date_raw).date()) if date_raw else None
    except Exception:
        date = date_raw

    doc_number = t(cfg['cols']['doc_number'])
    parties = t(cfg['cols']['parties'])
    address = t(cfg['cols']['address'])
    link = a(cfg['cols']['doc_link'])
    if link and not link.startswith('http'):
        link = urljoin(cfg['base_url'], link)

    return {
        "event_type": "lis_pendens",
        "event_date": date,
        "doc_number": doc_number,
        "parties": parties,
        "address": address,
        "doc_url": link
    }

def scrape_county(cfg, max_pages=3, delay=1.0):

    page = 1
    rows_out = []
    while page <= max_pages:
        url = cfg['start_url'].replace(cfg['page_param'] + "=1", f"{cfg['page_param']}={page}")
        try:
            html = fetch_page(url)
        except Exception as e:
            print("Fetch error:", e)
            break
        soup = BeautifulSoup(html, "lxml")
        rows = soup.select(cfg['results_selector'])
        if not rows:
            print("No rows found on page", page)
            break
        for r in rows:
            parsed = parse_row(r, cfg)
            rows_out.append(parsed)
        page += 1
        time.sleep(delay)
    return rows_out
```
